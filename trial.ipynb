```python
import numpy as np
from kev_plot import Plot

# Create sample data
x = np.linspace(0, 10, 100)
y = np.random.random_sample(100)
noise = np.random.normal(0, 0.1, size=100)
y_noisy = y + noise

categories = ['A', 'B', 'C', 'D']
heights = [10, 15, 7, 12]

data = np.random.normal(loc=0, scale=1, size=1000)

# Create plotter instance
plotter = Plot()

# Histogram with arguments
plotter.histogram(
    x=data,
    bins=30,
    color='skyblue',
    edgecolor='gray',
    alpha=0.8,
    histtype='stepfilled',
    title='Customized Histogram',
    xlabel='Random Value',
    ylabel='Density',
    density=True,
    grid=True,
    label='Sample Data',
    label_fontsize=14,
    title_fontsize=16,
    figsize=(10, 6),
    show=False
)

# Line plot with arguments
plotter.line(
    x=x,
    y=y_noisy,
    color='red',
    linestyle='--',
    linewidth=2.5,
    title='Line Plot',
    xlabel='X',
    ylabel='Y',
    label='line plot',
    grid=True,
    label_fontsize=14,
    title_fontsize=16,
    figsize=(12, 6),
    show=False
)

# Scatter plot with arguments
plotter.scatter(
    x=x,
    y=y,
    color='green',
    alpha=0.6,
    edgecolor='black',
    label='scatter ',
    title='Scatter plot)',
    xlabel='X',
    ylabel='y',
    label_fontsize=14,
    title_fontsize=16,
    figsize=(12, 6),
    show=False
)

# Bar plot arguments
plotter.bar(
    x=categories,
    height=heights,
    color='orange',
    alpha=0.7,
    edgecolor='black',
    title='Category Comparison',
    xlabel='Category',
    ylabel='Value',
    label='Values',
    grid=True,
    label_fontsize=14,
    title_fontsize=16,
    figsize=(10, 6),
    show=True
)
```
